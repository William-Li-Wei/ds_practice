```python
# During Coronavirus, how many confirmed case shall a small supermarket expect every day?
# Given the local effected rate = 0.001 and Supermarket's daily volume = 60000
n, p = 6000, 0.001
bi_data = np.random.binomial(n, p, 1000)
```
