# PDF, PMF and CDF

- **PDF**, Probability Density Function, is the density of a **continuous random variable**. 
- **PMF**, Probability Mass Function, is a function that gives the **probability** that a **discrete random variable** is exactly **equal to some value**.
- **CDF**, is the **probability** that **X** will take a value **less than or equal to x**.

```python
import numpy as np
import seaborn as sns

data = np.random.randn(500)
```

```python
# PDF
sns.distplot(data)
```

```python
# CDF
kwargs = {'cumulative': True}
sns.distplot(data, hist_kws=kwargs, kde_kws=kwargs)
```

# The first 4 moments of distribution

1. Expectation, `E(X) = mean(X)`
2. Variance, `Var(X) = E(X - E(X))^2`, tells how reliable the mean is.
3. Skewness, `Skew(X) = E(X - E(X))^3`, Negative if more extream values less than mean, Possitive if more extream values greater than mean.
4. Kurtosis, `Kurt(X) = E(X - E(X))^4`, is related to the tails of the distribution. Higher Kurtosis means more extream values.

### Mean tells where the distribution is along the axis.

### Variance, Skewness and Kurtosis tell the shape of the distribution.
